### - Task 3: Identify and handle missing values. Explain the chosen method for handling them.

```python
print("\nMissing values in each column:")
missing_values = df.isnull().sum()
print(missing_values)
```

```python
print(missing_values.sum())
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype in ['int64', 'float64']:
            df[col].fillna(df[col].median(), inplace=True)
            print(f"Filled missing values in column '{col}' with the median.")
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)
            print(f"Filled missing values in column '{col}' with the mode.")
```

```python
print("\nMissing values after handling:")
print(df.isnull().sum())
```

#### After checking for missing values, I found a few columns with null entries. I filled:
- Numerical columns with the **median** to prevent skew from outliers.
- Categorical columns with the **mode**, assuming the most frequent category is a reasonable replacement.

This ensures that no data is dropped unnecessarily while maintaining dataset integrity.

```python
# Option 2: Normalization (Min-Max scaling)
# Recommended when the data does not follow a Gaussian distribution,
# or when algorithms sensitive to the magnitude of features are used (e.g., K-Nearest Neighbors, Neural Networks).
# It scales features to a fixed range, usually [0, 1].
# If you prefer Min-Max scaling, uncomment the following lines and comment out the Standardization part.
# print("\nApplying Normalization (MinMaxScaler)...")
# scaler_minmax = MinMaxScaler()
# df[cols_to_scale] = scaler_minmax.fit_transform(df[cols_to_scale])
# print("DataFrame after Normalization:")
# print(df.head())
```

#### Summary
- Numerical features often need scaling to ensure that no single feature dominates the learning process due to its scale. I have chosen to apply **Standardization** (Z-score normalization) using `StandardScaler`. This transforms the features to have a mean of 0 and a standard deviation of 1.
- This is generally suitable for many machine learning algorithms, especially those that use distance metrics or assume normally distributed data.
- Columns resulting from one-hot encoding (binary columns) are excluded from scaling as they are already on a fixed scale (0 or 1).

```python
if task_type == "regression":
    print("\nTuning Random Forest Regressor...")
    model_to_tune = RandomForestRegressor(random_state=42)

    # Define the parameter distribution for RandomizedSearchCV
    # This is a more efficient option than GridSearchCV for large search spaces
    param_distributions = {
        'n_estimators': randint(50, 200), # Number of trees in the forest
        'max_depth': randint(3, 20),      # Maximum depth of the tree
        'min_samples_split': randint(2, 20), # Minimum number of samples required to split an internal node
        'min_samples_leaf': randint(1, 20),  # Minimum number of samples required to be at a leaf node
        'max_features': ['auto', 'sqrt', 'log2', None] # Number of features to consider when looking for the best split
    }

    # Initialize RandomizedSearchCV
    # n_iter: Number of parameter settings that are sampled. Higher = more thorough search, longer time.
    # cv: Number of folds for cross-validation.
    # scoring: Metric to optimize (e.g., 'neg_mean_squared_error' for regression, 'accuracy' or 'f1_weighted' for classification).
    # random_state: for reproducibility
    # n_jobs: Number of CPU cores to use (-1 means use all available cores)
    random_search = RandomizedSearchCV(estimator=model_to_tune,
                                       param_distributions=param_distributions,
                                       n_iter=50, # Number of iterations (adjust based on time budget)
                                       cv=5, # 5-fold cross-validation
                                       scoring='neg_mean_squared_error', # Optimize for lower MSE (higher negative MSE)
                                       random_state=42,
                                       n_jobs=-1,
                                       verbose=1) # Show progress

    # Fit RandomizedSearchCV to the training data
    random_search.fit(X_train, y_train)

    print("\nRandomizedSearchCV complete.")
    print("Best parameters found:", random_search.best_params_)
    print(f"Best cross-validation score (Negative MSE): {random_search.best_score_:.4f}")
    print(f"Best cross-validation score (RMSE): {np.sqrt(-random_search.best_score_):.4f}")


    # Get the best model found by RandomizedSearchCV
    best_model = random_search.best_estimator_

    # Evaluate the best model on the test set
    y_pred_tuned = best_model.predict(X_test)
    rmse_tuned = np.sqrt(mean_squared_error(y_test, y_pred_tuned))
    r2_tuned = r2_score(y_test, y_pred_tuned)

    print("\n--- Tuned Random Forest Regressor Performance on Test Set ---")
    print(f"RMSE: {rmse_tuned:.4f}")
    print(f"R2 Score: {r2_tuned:.4f}")

    # Compare with initial Random Forest performance from Task 15
    if 'Random Forest Regressor' in results:
         print("\nComparison with Initial Random Forest:")
         print(f"Initial RF Regressor - RMSE: {results['Random Forest Regressor']['RMSE']:.4f}, R2: {results['Random Forest Regressor']['R2']:.4f}")
         print(f"Tuned RF Regressor     - RMSE: {rmse_tuned:.4f}, R2: {r2_tuned:.4f}")
    else:
        print("\nNote: Initial Random Forest results not available for direct comparison.")


elif task_type == "classification":
    print("\nTuning Random Forest Classifier...")
    model_to_tune = RandomForestClassifier(random_state=42)

    # Define the parameter grid for GridSearchCV
    # GridSearchCV exhaustively searches over specified parameter values.
    # Can be computationally expensive for many parameters or large ranges.
    # Let's use GridSearchCV for classification for demonstration, but you can
    # switch to RandomizedSearchCV if the search space is large.
    param_grid = {
        'n_estimators': [50, 100, 150], # Fewer options for faster example
        'max_depth': [5, 10, 15],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
        # 'max_features': ['sqrt', 'log2'] # Add if needed
    }

    # Initialize GridSearchCV
    grid_search = GridSearchCV(estimator=model_to_tune,
                               param_grid=param_grid,
                               cv=5, # 5-fold cross-validation
                               scoring='f1_weighted', # Optimize for F1 Score (weighted)
                               n_jobs=-1,
                               verbose=1) # Show progress

    # Fit GridSearchCV to the training data
    grid_search.fit(X_train, y_train)

    print("\nGridSearchCV complete.")
    print("Best parameters found:", grid_search.best_params_)
    print(f"Best cross-validation score (Weighted F1): {grid_search.best_score_:.4f}")

    # Get the best model found by GridSearchCV
    best_model = grid_search.best_estimator_

    # Evaluate the best model on the test set
    y_pred_tuned = best_model.predict(X_test)
    accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
    f1_tuned = f1_score(y_test, y_pred_tuned, average='weighted')

    print("\n--- Tuned Random Forest Classifier Performance on Test Set ---")
    print(f"Accuracy: {accuracy_tuned:.4f}")
    print(f"F1-Score (Weighted): {f1_tuned:.4f}")

     # Compare with initial Random Forest performance from Task 15
    if 'Random Forest Classifier' in results:
         print("\nComparison with Initial Random Forest:")
         print(f"Initial RF Classifier - Accuracy: {results['Random Forest Classifier']['Accuracy']:.4f}, F1 Score: {results['Random Forest Classifier']['F1 Score']:.4f}")
         print(f"Tuned RF Classifier     - Accuracy: {accuracy_tuned:.4f}, F1 Score: {f1_tuned:.4f}")
    else:
        print("\nNote: Initial Random Forest results not available for direct comparison.")


else:
    print("\nHyperparameter tuning skipped: Task type not determined (regression or classification).")
```
